##### Is there a correlation between the season and the types of items purchased?

```python
# Cross-Tabulation:
# Create a cross-tabulation (also known as a contingency table) to display the frequency distribution of item categories for each season.

cross_tab = pd.crosstab(shopping_trends['Season'], shopping_trends['Item Purchased'])
print("Cross-Tabulation of Season and Item Purchased:")
print(cross_tab)
```

```python
#Visualization:
# Create a heatmap to visualize the relationship between seasons and the types of items purchased.

plt.figure(figsize=(12, 8))
sns.heatmap(cross_tab, cmap="YlGnBu", annot=True, fmt='d', cbar_kws={'label': 'Frequency'})
plt.title('Correlation between Season and Item Purchased')
plt.xlabel('Item Purchased')
plt.ylabel('Season')
plt.show()
```

One weird thing noted from above is that sunglasses are bought more in the winter time than in any other season.

```python
# Statistical Tests (Optional):
# Perform statistical tests, such as the Chi-square test, to assess the independence between seasons and types of items purchased.

from scipy.stats import chi2_contingency

# Chi-square test
chi2_stat, p_value, _, _ = chi2_contingency(cross_tab)
print("Chi-square Test Results:")
print(f"Chi2 Stat: {chi2_stat}, p-value: {p_value}")
```

The Chi-square test results indicate a Chi-square statistic of approximately 78.192 with a p-value of 0.2887. With a significance level of 0.05, we fail to reject the null hypothesis. This suggests that there is no significant association between seasons and types of items purchased. In other words, the observed distribution of items across seasons is consistent with what we would expect by random chance.

##### Do subscribed customers tend to make larger purchases?

```python
# Calculate basic descriptive statistics for purchase amounts of subscribed and non-subscribed customers separately.

subscribed_stats = shopping_trends[shopping_trends['Subscription Status'] == 'Yes']['Purchase Amount (USD)'].describe()
non_subscribed_stats = shopping_trends[shopping_trends['Subscription Status'] == 'No']['Purchase Amount (USD)'].describe()

print("Descriptive Statistics for Purchase Amounts of Subscribed Customers:")
print(subscribed_stats)
print("\nDescriptive Statistics for Purchase Amounts of Non-Subscribed Customers:")
print(non_subscribed_stats)
```

Non-subscribed customers make slightly larger purchases.

```python
# Create box plots or histograms to compare the distributions of purchase amounts between subscribed and non-subscribed customers.

plt.figure(figsize=(10, 6))
sns.boxplot(x='Subscription Status', y='Purchase Amount (USD)', data=shopping_trends, palette='Set2')
plt.title('Purchase Amounts by Subscription Status')
plt.xlabel('Subscription Status')
plt.ylabel('Purchase Amount (USD)')
plt.show()
```

Basically the same.

```python
# Perform a hypothesis test (e.g., t-test) to determine if there is a significant difference in purchase amounts between subscribed and non-subscribed customers.

from scipy.stats import ttest_ind

subscribed_purchase_amounts = shopping_trends[shopping_trends['Subscription Status'] == 'Yes']['Purchase Amount (USD)']
non_subscribed_purchase_amounts = shopping_trends[shopping_trends['Subscription Status'] == 'No']['Purchase Amount (USD)']

t_statistic, p_value = ttest_ind(subscribed_purchase_amounts, non_subscribed_purchase_amounts)
print("T-test Results:")
print("T-statistic:", t_statistic)
print("P-value:", p_value)
```

The t-statistic measures the difference between the means of two groups relative to the variation within the groups.
A negative t-statistic suggests that the mean purchase amount of subscribed customers is lower than that of non-subscribed customers, although the magnitude of this difference is small.
However, the absolute value of the t-statistic is relatively low, indicating that the difference between the means is not substantial.

The p-value is the probability of observing a t-statistic as extreme as the one calculated, assuming that there is no difference in purchase amounts between subscribed and non-subscribed customers.
In this case, the p-value of approximately 0.662 is relatively high, indicating that there is a 66.2% chance of observing the observed difference in means if there is actually no difference between the groups.
With a high p-value of 0.662, we fail to reject the null hypothesis, indicating that there is insufficient evidence to conclude that subscribed customers tend to make larger purchases compared to non-subscribed customers.

In summary, the results suggest that subscription status is not a significant predictor of purchase amounts, as there is no strong evidence of a difference in spending behavior between subscribed and non-subscribed customers.
